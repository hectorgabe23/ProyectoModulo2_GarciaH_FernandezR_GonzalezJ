```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

# Proyecto Simulación MonteCarlo
Se quiere demostrar cómo es que los Casinos ganan dinero.
Podríamos decir que el truco está en que cuanto mayor número de juegos, se incrementa la probabilidad de perder dinero.
Simularemos un dado de 100 caras en donde la casa tiene el 51% de ventaja, es decir si al rodar el dado tenemos un número igual a 51 o menos, la casa gana la apuesta, caso contrario el jugador gana. Esto le da a la casa un 2% de ventaja sobre el jugador.
<img src="https://miro.medium.com/max/894/1*J2XZDiZBxiHR5BKTDaX5ZQ.gif">


#### Definimos la función que simula el rodar del dado

```python
def rodar():
    
    dado = random.randint(1,100)
    
    if dado <=51:
        return False
    elif dado >51 & dado <=100:
        return True
```

### Definimos una función que nos simula el juego

Se toman en cuenta 3 argumentos principales:
 - Total de dinero con el que se inicia
 - Cuanto se va a apostar por juego
 - Total de Juegos
 
El función jugar nos regresa los fondos finales con los que el jugador se queda después de X cantidad de juegos. Además de graficar los juegos a través del tiempo vs los fondos después de cada apuesta.

```python
#1. total_dinero = Dinero con el que el jugador comienza
#2. apuesta = Dinero que se apuesta
#3. total_juegos = total de juegos
def jugar(total_dinero, apuesta, total_juegos):
    
    #Creamos listas vacías para:
    # 1.juego actual 
    # 2.dinero disponible
    # 3.última cantidad de dinero
    juego_act = []
    disponible = []
#empezamos con el número de juego 1
# si el número de juegos es menor que el máximo número de juegos, se sigue la simulación
    while juego < total_juegos:
        #si se gana
        if rodar():
            #agragamos el dinero a nuestros fondos
            total_dinero = total_dinero + apuesta
            #agragamos el número de juego en el que estamos
            juego_act.append(juego)
            #agragamos los nuevos fondos
            disponible.append(total_dinero)
        #si la casa gana
        else:
            #quitamos el dinero de nuestros fondos
            total_dinero = total_dinero - apuesta 
            #agregamos el número de juego
            juego_act.append(juego)
            #Agregamos el nuevo fondo
            disponible.append(total_dinero)
            
        #Incrementamos el número de juego por 1
        juego = juego + 1
        
    #gráfica de nuestros fondos através del tiempo
    plt.plot(juego_act,disponible)
    fondos_finales.append(disponible[-1])
    return(fondos_finales)
```

# Simulación
Haremos una simulación de 100 esecenarios posibles, guardaremos los resultados finales en una lista y usaremos el promedio para determinar cuál será el resultado final con mayor probabilidad

```python
#Llama a la función para simular los juegos y calcular los fondos restantes del jugador después de todas las apuestas
#Inicializamos el escenario en 1
x=1
#Lista de fondos finales
fondos_finales= []
juegos = 50
apuest = 500
fond_iniciales = 10000

while x<=100:
    fondo_final = jugar(fond_iniciales,apuest,juegos) # aquí se puede jugar con los parámetros(dinero_inicial, apuesta, juegos a jugar)
    x=x+1

#graficamos la línea de valor en cuenta vs número de juegos
plt.ylabel('Dinero del jugador $')
plt.xlabel('Número de apuestas')
plt.show()


#imprimimos el dinero con el que termina el jugador
print("El jugador empezó con $10,000 y termina con ${}".format(str(sum(fondo_final)/len(fondo_final))))
```

```python
plt.xlabel('Dinero Final $')
plt.ylabel('Probabilidad')
plt.title('Distribuición de fondos finales después de {} juegos apostando ${}'.format(juegos, apuest))
a, b, c = plt.hist(fondos_finales)
plt.hist(fondos_finales)
plt.grid(True)
plt.text(min(fondos_finales), max(a), r'$\mu={:.2f},\ \sigma={:.2f}$'.format(np.mean(fondos_finales), np.std(fondos_finales)))
plt.tight_layout()
plt.show()
```
